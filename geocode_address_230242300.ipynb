```python
# import relevant libraries
import pandas as pd
from arcgis.gis import GIS
from arcgis.geocoding import geocode

# path to the danish vineyards point layer, in postcode format
vineyards = pd.read_csv(r'U:\Modules\TRP483\data\vineyards\Danish_Vineyards.csv')

gis = GIS("home")

# Function which extracts longitude and latitude coordinates from 'address', 'town' and 'ZIP' columns
def geocode_address(row):
    address = f"{row['address']}, {row['town']}, {row['zip']}, Denmark"
    result = geocode(address, as_featureset=True)
    if result and result.features:
        location = result.features[0].geometry
        return location['x'], location['y']
    else:
        return None, None

# Create new columns and apply geocode_address function to dataframe 'vineyards'
vineyards[['Longitude', 'Latitude']] = vineyards.apply(geocode_address, axis=1, result_type="expand")

# Remove NA values if present
vineyards = vineyards.dropna(subset=['Longitude', 'Latitude'])

# ouput the new geocoded layer
output_path = r'U:\Modules\TRP483\data\vineyards\Danish_Vineyards_decoded.csv'
vineyards.to_csv(output_path, index=False)

```
